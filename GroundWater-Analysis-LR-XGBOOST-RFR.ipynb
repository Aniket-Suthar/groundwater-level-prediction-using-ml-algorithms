```python
# Identifyin null values
null_values = data.isnull().sum()
print("Null Values:")
print(null_values)
```

*5. UPDATING FEATURES HAVING VALUE 0*

```python
# Identify 0 values
zero_values = (data == 0).sum()
print("\nZero Values:")
print(zero_values)
```

```python
# Identify numeric columns (excluding 'Year' and 'States')
numeric_columns = [col for col in data.columns if col not in ['  Year', 'States']]

# Calculating mean of each numeric column (excluding 'Year' and 'States')
mean_values = data[numeric_columns].median()

# Replacing zero values with mean for each numeric column (excluding 'Year' and 'States')
for col in numeric_columns:
    data[col] = data[col].replace(0, mean_values[col])
```

*7. REMOVING NON-USEFUL COLUMNS*

```python
# Separate features (X) and target variables (y)
X = data.drop(['Min Depth to Water Level(mbgl)', 'Max Depth to Water Level','  Year','States'],axis=1)
y_min = data['Min Depth to Water Level(mbgl)']
```

*8. DATA SPLITTING INTO TRAINING AND TESTING*

```python
# Split data into training and testing sets
X_train, X_test, y_train_min, y_test_min = train_test_split(X, y_min, test_size=0.2, random_state=30)
```

```python
type(y_train_min)
```

```python
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

# **LINEAR REGRESSION**

```python
# Standardize numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# X_train_scaled
```

```python
# X_test_scaled
```

```python
# Initialize and train linear regression models
regression_min = LinearRegression()
```

```python
regression_min.fit(X_train_scaled, y_train_min)
```

```python
# Predictions
y_min_pred_lr = regression_min.predict(X_test_scaled)
```

*A. EVALUATION METRICS*

```python
#metrics for test data

print("The explained variance score for min depth: ",explained_variance_score(y_test_min, y_min_pred_lr))
print("The r2 score for min depth: ",r2_score(y_test_min, y_min_pred_lr))
print("The mean square error for min depth: ",mean_squared_error(y_test_min, y_min_pred_lr))
print("The mean absolute error for min depth: ",mean_absolute_error(y_test_min, y_min_pred_lr))
print("The root mean square error for min depth: ",root_mean_squared_error(y_test_min,y_min_pred_lr))
```

```python
#metrics for the training set
y_train_pred = regression_min.predict(X_train_scaled)

print("The explained variance score for min depth: ",explained_variance_score(y_train_min, y_train_pred))
print("The r2 score for min depth: ",r2_score(y_train_min, y_train_pred))
print("The mean square error for min depth: ",mean_squared_error(y_train_min, y_train_pred))
print("The mean absolute error for min depth: ",mean_absolute_error(y_train_min, y_train_pred))
print("The root mean square error for min depth: ",root_mean_squared_error(y_train_min, y_train_pred))
```
